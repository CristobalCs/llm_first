```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Usando dispositivo: {device}")
# Esta línea verifica que tu RTX 3050 está siendo utilizada
print(f"Nombre de la GPU: {torch.cuda.get_device_name(0)}") 

celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=np.float32)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=np.float32)

entradas = torch.from_numpy(celsius).unsqueeze(1).to(device)
salidas_reales = torch.from_numpy(fahrenheit).unsqueeze(1).to(device)
```

```python
class RegresionLineal(nn.Module):
    def __init__(self):
        super(RegresionLineal, self).__init__()
        self.capa_lineal = nn.Linear(1, 1) 

    def forward(self, x):
        return self.capa_lineal(x)

modelo = RegresionLineal().to(device)
criterio = nn.MSELoss() 
optimizador = optim.Adam(modelo.parameters(), lr=0.1) 

historial_perdida = []
num_epocas = 1000 

for epoch in range(num_epocas):
    salidas_predichas = modelo(entradas)
    perdida = criterio(salidas_predichas, salidas_reales)
    historial_perdida.append(perdida.item()) 
    optimizador.zero_grad() 
    perdida.backward()      
    optimizador.step()      

    if (epoch + 1) % 100 == 0: 
        print(f'Época [{epoch+1}/{num_epocas}], Pérdida: {perdida.item():.4f}')
```
